```python
import os
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import re
```

```python
# The source table is a general output from the sort_and_extract
# script, where context and occurrence columns were removed,
# and the Partie/Chapitre/Section columns were reduced in one
# "Localisation" column. Changes were made directly in the output file
# this time, but visible in the dataviz script.
author_orig = pd.read_csv(
    "/output/2022-09-29_full-output_avec-themes.csv",
    skip_blank_lines=False,
    infer_datetime_format = True,
    na_filter=False
)
print(list(author_orig))
        
author_orig.head(10)
```

```python
locs_orig = sorted(author_orig['Localisation'].unique())
locs = {}



print(len(locs_orig))
    
for item in locs_orig:
    lowitem = item.lower()
    if item not in locs.keys():
        if re.search(r"do[vuü]aire", lowitem):
            locs[item] = "douaire"
        elif re.search(r"(testamen)|(patrimoine)|(s[vu]ccession)|(donation)|(inventaire)|(servit[vu]de)|(heritage)", lowitem):
            locs[item] = "successions"
        elif "<front>" in lowitem:
            locs[item] = "matière liminaire"
        elif re.search(r"(fief)|(personne)|([ée]glise)|(seigneur)|(ro[iy])", lowitem):
            locs[item] = "société"
        elif re.search(r"(crime)|(garde)|(q[vu]erelle)|(execution)|(d[ée]cret)|(haro)|([jil]ustice)|([jil]uge)|([jil][uv]ris?diction)|(st[iy]l+e)", lowitem):
            locs[item] = "justice"
        elif "mariage" in lowitem:
            locs[item] = "mariage"
        elif re.search(r"(meuble)|(varech)|(banon)|(fores?t)|(prescription)|(bourse)|(namp)", lowitem):
            locs[item] = "propriété"
        else:
            print(item)
            
author_orig['Thème'] = ''

for idx, row in author_orig.iterrows():
    row['Thème'] = locs[row['Localisation']]
    
author_orig['Thème'].head()

author_orig.to_csv("output/2022-09-29_full-output_avec-themes.csv")
```
