```python
import os
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import re
from rdflib import Graph, Literal, RDF, URIRef
from rdflib.namespace import FOAF , XSD
import lxml.etree as ET
import xmltodict
import json

ET.register_namespace("rdf", "http://www.w3.org/1999/02/22-rdf-syntax-ns#")
ET.register_namespace("rdfs", "http://www.w3.org/2000/01/rdf-schema#")

author_orig = pd.read_csv(
    "output/2022-09-29_full-output_avec-themes.csv",
    skip_blank_lines=False,
    infer_datetime_format = True,
    na_filter=False
)
print(list(author_orig))
```

```python
all_mentions = {
    'authors' : {},
    'locations' : {}
}

author_dict = {}
loc_dict = {}
done = []

entities = 0
predicates = 0

for author in author_orig['Auteur'].unique():
    entities += 1
    ident = "A" + str(entities)
    author_dict[author] = ident
    all_mentions['authors'][ident] = {
        'label': author,
        'mentionned in' : [],
        'along with': {}
    }

entities = 0

for location in author_orig['Localisation'].unique():
    entities += 1
    locid = "L" + str(entities)
    loc_dict[location] = locid
    all_mentions['locations'][locid] = {
        'label': location,
        'mentions' : []
    }
        
for idx, row in author_orig.iterrows():
    author = row['Auteur']
    autid = author_dict[author]
    location = row['Localisation']
    locid = loc_dict[location]
    
    all_mentions['authors'][autid]['mentionned in'].append(locid)
                
for idx, row in author_orig.iterrows():
    if row['Localisation'] not in done:
        location = row['Localisation']
        locid = loc_dict[location]
        theme = row['Thème']

        all_mentions['locations'][locid]['has theme'] = row['Thème']
        done.append(row['Localisation'])

print(all_mentions)
```

```python

for aAuthor in all_mentions['authors'].keys():
    aMentions = all_mentions['authors'][aAuthor]['mentionned in']
    for bAuthor in all_mentions['authors'].keys():
        if aAuthor != bAuthor:
            bMentions = all_mentions['authors'][bAuthor]['mentionned in']
            for loc in bMentions:
                if loc in aMentions and bAuthor in all_mentions['authors'][aAuthor]['along with']:
                    all_mentions['authors'][aAuthor]['along with'][bAuthor] += 1
                elif loc in aMentions:
                    all_mentions['authors'][aAuthor]['along with'][bAuthor] = 1
for location in all_mentions['locations'].keys():
    for author in all_mentions['authors'].keys():
        if location in all_mentions['authors'][author]['mentionned in']:
            all_mentions['locations'][location]['mentions'].append(author)

            
"""author_csv = []
for ident in all_mentions['authors'].keys():
    author = all_mentions['authors'][ident]
    if len(author['along with'].keys()) != 0:
        for collegue in author['along-with'].keys().unique():
            line = {
                'Auteur' : author['label'],
                ''
            }
"""
print(all_mentions)
```

```python
couples = {}
out_couples = []

locsv = []

for location in all_mentions['locations'].keys():
    done = []
    thisloc = all_mentions['locations'][location]
    if len(all_mentions['locations'][location]['mentions']) > 1:
        mentionned = all_mentions['locations'][location]['mentions']
        for aAuthor in np.unique(mentionned):
            aName = all_mentions['authors'][aAuthor]['label']
            for bAuthor in np.unique(mentionned):
                bName = all_mentions['authors'][bAuthor]['label']
                couple = aName + " - " + bName
                inverse = bName + " - " + aName
                if aAuthor != bAuthor and couple not in done and inverse not in done:
                    locsv.append({
                        'couple' : couple,
                        'localisation' : thisloc['label'],
                        'theme' : thisloc['has theme']
                    })
                    try:
                        couples[couple][thisloc['has theme']] += 1
                    except:
                        try:
                            couples[couple][thisloc['has theme']] = 1
                        except:
                            couples[couple] = {thisloc['has theme'] : 1}
                    done.append(couple)

for couple in couples.keys():
    for theme in couples[couple].keys():
        if couples[couple][theme] > 2:
            out_couples.append({
                'couple':couple,
                'theme' : theme,
                'co-occurrences' : str(couples[couple][theme])
            })
                    
with open('dataviz/2022-09-28/couples-localisations.csv', 'w') as csvfile:
    csvwriting = csv.DictWriter(csvfile, fieldnames = locsv[0].keys())
    csvwriting.writeheader()
    csvwriting.writerows(locsv)
with open('dataviz/2022-09-28/couples-resume.csv', 'w') as csvfile:
    csvwriting = csv.DictWriter(csvfile, fieldnames = out_couples[0].keys())
    csvwriting.writeheader()
    csvwriting.writerows(out_couples)
```
